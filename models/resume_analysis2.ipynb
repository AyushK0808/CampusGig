```python
#converting .pdf to .txt

import PyPDF2

pdf_file_path="/content/machine-learning-engineer - Template 14 .pdf" #Change filename to input file
def extract_text_from_pdf(pdf_file_path):
    try:
        pdf_reader = PyPDF2.PdfReader(pdf_file_path)

        text = ''
        for page in pdf_reader.pages:
            text += page.extract_text()

        return text
    except Exception as e:
        print("An error occurred:", str(e))
        return None

# Replace 'example.pdf' with the path to your PDF file


extracted_text = extract_text_from_pdf(pdf_file_path)
if extracted_text:
    output_file_path = '/content/output.txt'  # Specify the path for the output text file
    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        output_file.write(extracted_text)
    print("Text extracted from PDF and saved to", output_file_path)
```

```python
#removing stopwords and punctuations from .txt

import nltk
import string
from nltk.corpus import stopwords

# Download NLTK stop words data
nltk.download('stopwords')

# Define the path to your input text file
input_text_file = '/content/output.txt'

# Read the text from the input file
with open(input_text_file, 'r') as file:
    text = file.read()

# Tokenize the text (split it into words)
words = nltk.word_tokenize(text)

# Remove punctuation and convert to lowercase
words = [word.lower() for word in words if word not in string.punctuation]

# Remove stop words
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]

# Join the filtered words back into a text string
filtered_text = ' '.join(filtered_words)

# Save the filtered text to a new file or print it
with open('/content/output.txt', 'w') as output_file:
    output_file.write(filtered_text)

# Print the filtered text
print(filtered_text)
```

```python
#pipeline for model

from transformers import pipeline

# Load the zero-shot classification pipeline with the desired model
# You can choose different models, like 'facebook/bart-large-mnli' or 'roberta-large-mnli'
classifier = pipeline("zero-shot-classification", device=0)

```

```python
import csv

# Example class descriptions and input text
class_descriptions= ['App Development','Frontend Development','Backend Development','UI-UX Design','AI/ML','Cybersecurity','Blockchain','DBMS']
input_text_file = '/content/output.txt'
```

```python
with open(input_text_file, 'r') as textfile:
    input_text = textfile.read()
```

```python

# Perform zero-shot classification
result = classifier(input_text, class_descriptions)
```

```python
# Print the results
#print("Input Text:", input_text)
print("Classifications:")
for label, score in zip(result["labels"], result["scores"]):
    print(f"{label}: {score:.3f}")
```
